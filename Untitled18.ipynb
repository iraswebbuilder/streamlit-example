```python
!pip install streamlit
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import numpy as np


# Mengatur tema latar belakang gelap dan palet warna berwarna-warni
sns.set(style='white', palette='deep')

```

```python
# Baca/buka file dataset yang digunakan
day = pd.read_csv('day.csv')
hour = pd.read_csv('hour.csv')

# Gabungkan dataset menjadi satu
df = pd.concat([day, hour], axis=0)

# Tampilkan data
display(df)
```

```python
#hapus data kosong/ data yang tidak diperlukan
df = df.drop(columns=['hr', 'instant', 'windspeed'])

#ubah/sesuaikan nama data
df.rename(columns={'yr': 'year', 'mnth': 'month', 'weathersit': 'weather', 'cnt': 'count', 'kode': 'new_name'}, inplace=True)

#mengubah angka menjadi keterangan
df['month'] = df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
df['season'] = df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
df['weekday'] = df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
df['weather'] = df['weather'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Menyiapkan daily_rent_df
def create_daily_rent_df(df):
    daily_rent_df = df.groupby(by='dteday').agg({
        'count': 'sum'
    }).reset_index()
    return daily_rent_df

# Menyiapkan daily_casual_rent_df
def create_daily_casual_rent_df(df):
    daily_casual_rent_df = df.groupby(by='dteday').agg({
        'casual': 'sum'
    }).reset_index()
    return daily_casual_rent_df

# Menyiapkan daily_registered_rent_df
def create_daily_registered_rent_df(df):
    daily_registered_rent_df = df.groupby(by='dteday').agg({
        'registered': 'sum'
    }).reset_index()
    return daily_registered_rent_df

# Menyiapkan season_rent_df
def create_season_rent_df(df):
    season_rent_df = df.groupby(by='season')[['registered', 'casual']].sum().reset_index()
    return season_rent_df
# Menyiapkan monthly_rent_df
def create_monthly_rent_df(df):
    monthly_rent_df = df.groupby(by='month').agg({
        'count': 'sum'
    })
    ordered_months = [
        'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
        'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
    ]
    monthly_rent_df = monthly_rent_df.reindex(ordered_months, fill_value=0)
    return monthly_rent_df

# Menyiapkan weekday_rent_df
def create_weekday_rent_df(df):
    weekday_rent_df = df.groupby(by='weekday').agg({
        'count': 'sum'
    }).reset_index()
    return weekday_rent_df

# Menyiapkan workingday_rent_df
def create_workingday_rent_df(df):
    workingday_rent_df = df.groupby(by='workingday').agg({
        'count': 'sum'
    }).reset_index()
    return workingday_rent_df

# Menyiapkan holiday_rent_df
def create_holiday_rent_df(df):
    holiday_rent_df = df.groupby(by='holiday').agg({
        'count': 'sum'
    }).reset_index()
    return holiday_rent_df

# Menyiapkan weather_rent_df
def create_weather_rent_df(df):
    weather_rent_df = df.groupby(by='weather').agg({
        'count': 'sum'
    })
    return weather_rent_df
```

```python
# Membuat komponen filter
min_date = pd.to_datetime(df['dteday']).dt.date.min()
max_date = pd.to_datetime(df['dteday']).dt.date.max()
```

```python
# Menyiapkan berbagai dataframe
daily_rent_df = create_daily_rent_df(df)
daily_casual_rent_df = create_daily_casual_rent_df(df)
daily_registered_rent_df = create_daily_registered_rent_df(df)
season_rent_df = create_season_rent_df(df)
monthly_rent_df = create_monthly_rent_df(df)
weekday_rent_df = create_weekday_rent_df(df)
workingday_rent_df = create_workingday_rent_df(df)
holiday_rent_df = create_holiday_rent_df(df)
weather_rent_df = create_weather_rent_df(df)
```

```python
# Membuat Dashboard secara lengkap

# Membuat judul
st.header('Bike Rental Dashboard')
```

```python
# Membuat jumlah penyewaan harian
st.subheader('Daily Rentals')
col1, col2, col3 = st.columns(3)

with col1:
    daily_rent_casual = daily_casual_rent_df['casual'].sum()
    st.metric('Casual User', value= daily_rent_casual)

with col2:
    daily_rent_registered = daily_registered_rent_df['registered'].sum()
    st.metric('Registered User', value= daily_rent_registered)

with col3:
    daily_rent_total = daily_rent_df['count'].sum()
    st.metric('Total User', value= daily_rent_total)
```

```python
plt.figure(figsize=(10, 5))  # Membuat gambar dengan ukuran 10x5

# Scatter plot untuk 'temp' vs 'count'
plt.subplot(1, 3, 1)  # Membuat subplot pertama dalam grid 1x3
sns.scatterplot(
    x='temp',
    y='count',
    data=df,
    alpha=0.5,  # Menambahkan transparansi pada titik-titik
    color='pink'  # Mengubah warna plot menjadi pink
)
plt.title('Temperature vs Count')  # Menambahkan judul

# Scatter plot untuk 'atemp' vs 'count'
plt.subplot(1, 3, 2)  # Membuat subplot kedua dalam grid 1x3
sns.scatterplot(
    x='atemp',
    y='count',
    data=df,
    alpha=0.5,  # Menambahkan transparansi pada titik-titik
    color='pink'  # Mengubah warna plot menjadi pink
)
plt.title('Feels Like Temperature vs Count')  # Menambahkan judul

# Scatter plot untuk 'hum' vs 'count'
plt.subplot(1, 3, 3)  # Membuat subplot ketiga dalam grid 1x3
sns.scatterplot(
    x='hum',
    y='count',
    data=df,
    alpha=0.5,  # Menambahkan transparansi pada titik-titik
    color='pink'  # Mengubah warna plot menjadi pink
)
plt.title('Humidity vs Count')  # Menambahkan judul

plt.tight_layout()  # Mengatur tata letak subplot agar rapi
plt.show()  # Menampilkan plot
```

```python
# Mengatur ukuran gambar lebih besar/Menggunakan Seaborn untuk membuat heatmap
fig, ax = plt.subplots(figsize=(10,5))
correlation_matrix = df.corr(numeric_only=True)
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
# Membuat heatmap dengan peta warna "rainbow"
sns.heatmap(
    correlation_matrix,
    annot=True,
    mask=mask,
    cmap="rainbow",
    center=0,
    fmt=".2f")
# Menambahkan judul
plt.title("Heatmap Korelasi antara Variabel")
# Menampilkan plot
plt.show()
```

```python
# Membuah jumlah penyewaan berdasarkan kondisi cuaca
st.subheader('Weatherly Rentals')

# Membuat subplot dengan 3 baris dan 1 kolom, dengan ukuran gambar 10x5
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 5))

# Barplot berdasarkan workingday
sns.barplot(
    x='workingday',
    y='count',
    data=df,
    ax=axes[0],
    palette='pink')
axes[0].set_title('Penyewa Sepeda Berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Penyewa Sepeda')

# Barplot berdasarkan holiday
sns.barplot(
    x='holiday',
    y='count',
    data=df,
    ax=axes[1],
    palette='pink')
axes[1].set_title('Penyewa Sepeda Berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Penyewa Sepeda')

# Barplot berdasarkan weekday
sns.barplot(
    x='weekday',
    y='count',
    data=df,
    ax=axes[2],
    palette='pink')
axes[2].set_title('Penyewa Sepeda Berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Penyewa Sepeda')

# Mengatur tata letak subplot agar sesuai
plt.tight_layout()

# Menampilkan plot
plt.show()
st.pyplot(fig)
```

```python
# Membuat jumlah penyewaan berdasarkan season
st.subheader('Seasonly Rentals')

# Mengubah kolom 'month' menjadi tipe data kategorikal dengan urutan tertentu
df['month'] = pd.Categorical(df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

# Menghitung jumlah total sepeda yang disewakan berdasarkan bulan dan tahun
monthly_counts = df.groupby(by=["month", "year"]).agg({
    "count": "sum"
}).reset_index()

# Membuat line plot (grafik garis) berdasarkan data jumlah sepeda yang disewakan
sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rainbow",  # Mengatur palet warna 'rainbow'
    marker="o"  # Menambahkan marker 'o' pada titik data
)

# Memberikan judul pada grafik
plt.title("Jumlah total sepeda yang disewakan berdasarkan Bulan dan tahun")

# Menghapus label sumbu x
plt.xlabel(None)

# Menghapus label sumbu y
plt.ylabel(None)

# Menampilkan legenda dengan judul 'Tahun' di lokasi kanan atas
plt.legend(title="Tahun", loc="upper right")

# Mengatur tata letak yang lebih baik
plt.tight_layout()

# Menampilkan grafik
plt.show()
st.pyplot(fig)
```
